# PROPERTIES OF ENZYME ACTION

### Purpose

Enzymes are complex proteins that start and accelerate specific chemical reactions without being altered or used. Enzymes act as catalyst which provide temporary binding sites for specific substrate molecules. During this lab, we are going to observe enzymes denature and lose its catalytic abilites. We will be learning the effect of pH and temperature on an enzyme.

### Procedures

3-C:Digestion of fat with pancreatic lipase and bile salts

Add just enough water litmus powder to a container of dairy cream to produce a medium blue color. Pour 3 ml of the litmus cream into 4 separate test tubes. Into two additional test tubes pour 3 ml of 2% pancreatin. Preincubate the litmus cream and the pancreatin separately in a 37 degree celcius water bath for 5 minutes. Then prepare four test tubes as follows:

Tube #1: 3 ml cream + 3 ml pancreatin

Tube #2: 3 ml cream + 3 ml distilled water

Tube #3: 3 ml cream + 3 ml pancreatin + pinch of bile salts

Tube #4: 3 ml cream + 3 ml distilled water + pinch of bile salts

Gently shake each tube for 30 seconds to mix in the bile salts. Incubate all four tubes in a 37 degree celcius water bath for 1 hour, checking every minute for the first 5 minutes or until the first tube changes color, then every 15 minutes for the rest of the hour. Record the time and number of the tube. Continue checking for the remainder of the hour. Remove the tubes from the water bath. Test the pH of each tube using pH paper and note the odor and color of each tube. Summarize results. Explain how the digestion of fat affects the pH of the solution and how bile affects the rate of digestion.

### Discussion

The table above displays the results from the experiment conducted. From this we can observe the color change in the solution as well as the pH change in each testing tube. They all started to change color at around the same time which was at the 5 minute mark. We are also able to see that test tube #3 which had both pancreatin and bile salts had the great change in pH. You can also note that testing tube #2 which had no assistance with distilled water hardly changed in color and the pH level. We are able to makeout the role that pancreatin and bile salts have on the digestion of fats and the effect it has on the pH.

### Conclusion 

Enzymes are proteins that act as biological catalysts by accelerating chemical reactions. The purpose of this lab was to see role that temperature has on enzymes and the effects of pH. The pH changes the enzyme activity. Molecules move faster at higher temperatures. The faster movement of molecules increases the chance of collision, thus making the binding between substrate and enzyme more likely.
